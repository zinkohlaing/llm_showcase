```python
# Create empty meeting audio
full_meeting = AudioSegment.silent(duration=1000)
```

```python
# Generate each line
for i, (speaker, text) in enumerate(meeting_script):
    tts = gTTS(text=text, lang='en', slow=False)
    filename = f"temp_line_{i}.mp3"
    tts.save(filename)

    segment = AudioSegment.from_file(filename, format="mp3")
    full_meeting += segment + AudioSegment.silent(duration=700)

# Export the final audio
full_meeting.export("meeting_audio.mp3", format="mp3")
```

```python
# Cleanup
for i in range(len(meeting_script)):
    os.remove(f"temp_line_{i}.mp3")

print("✅ Meeting audio ready: meeting_audio.mp3")
```

```python
# imports

import os
import requests
from IPython.display import Markdown, display, update_display
from openai import OpenAI
from google.colab import drive
from huggingface_hub import login
from google.colab import userdata
from transformers import AutoTokenizer, AutoModelForCausalLM, TextStreamer, BitsAndBytesConfig
import torch
```

```python
# Constants - for models

AUDIO_MODEL = 'whisper-1'
LLAMA = "meta-llama/Meta-Llama-3.1-8B-Instruct"
```

```python
audio_filename = "/content/drive/MyDrive/Education/Udemy/LLM_Engineering_Course/meeting_audio.mp3"
```

```python
# Sign in to HuggingFace Hub

hf_token = userdata.get('HF_TOKEN')
login(hf_token, add_to_git_credential=True)
```

```python
# Sign in to OpenAI using Secrets in Colab

openai_api_key = userdata.get('OPENAI_API_KEY')
openai = OpenAI(api_key=openai_api_key)
```

```python
# Use the Whisper OpenAI model to convert the Audio to Text
# If you'd prefer to use an Open Source model, class student Youssef has contributed an open source version
# which I've added to the bottom of this colab
# rb stands for read mode (file is opened for reading and binary mode, file should be treated as binary data)
audio_file = open(audio_filename, "rb")
transcription = openai.audio.transcriptions.create(model=AUDIO_MODEL, file=audio_file, response_format="text")
print(transcription)
```

```python
system_message = "You are an assistant that produces minutes of meetings from transcripts, with summary, key discussion points, takeaways and action items with owners, in markdown."
user_prompt = f"Below is an extract transcript of an internal meeting. Please write minutes in markdown, including a summary with attendees, location and date; discussion points; takeaways; and action items with owners.\n{transcription}"

messages = [
    {"role": "system", "content": system_message},
    {"role": "user", "content": user_prompt}
  ]
```

```python
# quantizing to reduce the memory usage
quant_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_quant_type="nf4"
)
```

```python
tokenizer = AutoTokenizer.from_pretrained(LLAMA)
tokenizer.pad_token = tokenizer.eos_token
inputs = tokenizer.apply_chat_template(messages, return_tensors="pt").to("cuda")
streamer = TextStreamer(tokenizer)
model = AutoModelForCausalLM.from_pretrained(LLAMA, device_map="auto", quantization_config=quant_config)
outputs = model.generate(inputs, max_new_tokens=2000, streamer=streamer)
```

```python
response = tokenizer.decode(outputs[0])
```

```python
import gradio as gr
import torch
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig

# Load Whisper for transcription
asr = pipeline(
    "automatic-speech-recognition",
    model="distil-whisper/distil-large-v3",  # Fast and accurate
    chunk_length_s=30,
    device=0 if torch.cuda.is_available() else "cpu"
)

# Load Llama for summarization
LLAMA_MODEL = "meta-llama/Meta-Llama-3.1-8B-Instruct"
quant_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_quant_type="nf4"
)
tokenizer = AutoTokenizer.from_pretrained(LLAMA_MODEL)
model = AutoModelForCausalLM.from_pretrained(LLAMA_MODEL, device_map="auto", quantization_config=quant_config)

def transcribe(audio_path):
    # Transcribe audio to text
    result = asr(audio_path)
    return result["text"]

def summarize(transcript):
    # Format prompt for meeting minutes
    system_message = "You are an assistant that produces minutes of meetings from transcripts, with summary, key discussion points, takeaways and action items with owners, in markdown."
    user_prompt = f"Below is an extract transcript of a meeting. Please write minutes in markdown, including a summary with attendees, location and date; discussion points; takeaways; and action items with owners.\n{transcript}"
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": user_prompt}
    ]
    input_ids = tokenizer.apply_chat_template(messages, return_tensors="pt").to(model.device)
    output = model.generate(input_ids, max_new_tokens=1024)
    summary = tokenizer.decode(output[0], skip_special_tokens=True)
    return summary

def process_audio(audio_path):
    transcript = transcribe(audio_path)
    summary = summarize(transcript)
    return summary
```

```python
iface = gr.Interface(
    fn=process_audio,
    inputs=gr.Audio(type="filepath", label="Upload Audio File"),
    outputs=gr.Textbox(label="Meeting Minutes and Key Points"),
    title="Meeting Minutes Generator from Audio",
    description="Upload an audio file (e.g., from GTTS). The app will transcribe and summarize the meeting."
)
iface.launch(share=True)
```
